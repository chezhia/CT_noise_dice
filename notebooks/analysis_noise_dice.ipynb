```python
import os
import numpy as np
import pandas as pd
from scipy.stats import pearsonr
import statsmodels.api as sm
import plotly.graph_objects as go

# ==========================
# Define Parameters
# ==========================

# Noise metric to plot against Dice scores
#selected_noise = 'original_firstorder_Uniformity'  # Ensure this column exists in your DataFrame
selected_noise = 'liver_noise'  # Ensure this column exists in your DataFrame

# Define the two models and their corresponding Dice metric columns
models = {
    'TotalSegmentator': 'TS_macro_avg_dice',  # Replace with your actual column names
    'Vista3D': 'macro_avg_dice'                     # Replace with your actual column names
}

# # # # Define the two models and their corresponding Dice metric columns
# models = {
#     'TotalSegmentator': 'TS_dice_bladder',  # Replace with your actual column names
#     'Vista3D': 'dice_bladder'                     # Replace with your actual column names
# }

# Define colors for the models
colors = {
    'TotalSegmentator': 'blue',
    'Vista3D': 'green'
}

# Define the output directory
output_dir = 'figures'
os.makedirs(output_dir, exist_ok=True)

# ==========================
# Plotting Function
# ==========================
def plot_two_models_noise_vs_dice(df, selected_noise, models, colors, output_dir, percentile_range=(0, 98)):
    """
    Plots Dice scores vs noise metrics for two AI models on the same scatter plot.
    Each model's data points are in different colors with separate trendlines and annotations.
    Outliers in the noise metric are excluded based on the specified percentile range.
    
    Parameters:
    - df (DataFrame): The input DataFrame containing noise metrics and Dice scores.
    - selected_noise (str): The column name for the noise metric.
    - models (dict): Dictionary mapping model names to their Dice score column names.
    - colors (dict): Dictionary mapping model names to colors.
    - output_dir (str): Directory to save the output HTML plots.
    - percentile_range (tuple): Tuple indicating the lower and upper percentiles for outlier exclusion.
    """
    # Initialize a Plotly figure
    fig = go.Figure()
    
    # Initialize list for annotations
    annotations = []
    
    for idx, (model_name, dice_col) in enumerate(models.items(), start=1):
        # Check if the Dice score column exists
        if dice_col not in df.columns:
            print(f"Column '{dice_col}' for model '{model_name}' not found. Skipping this model.")
            continue
        
        # Filter out rows with NaNs in noise or dice metric
        subset_df = df[[selected_noise, dice_col, 'ct_name']].dropna()
        
        if subset_df.empty:
            print(f"No data available for model '{model_name}'. Skipping.")
            continue
        
        # Percentile-Based Filtering to exclude outliers
        lower_percentile, upper_percentile = percentile_range
        lower_bound = np.percentile(subset_df[selected_noise], lower_percentile)
        upper_bound = np.percentile(subset_df[selected_noise], upper_percentile)
        
        # Filter the DataFrame to include only the central 90%
        filtered_df = subset_df[
            (subset_df[selected_noise] >= lower_bound) &
            (subset_df[selected_noise] <= upper_bound)
        ]
        
        if filtered_df.empty:
            print(f"All data for model '{model_name}' excluded after outlier removal. Skipping.")
            continue
        
        # Compute Pearson correlation coefficient and p-value
        r_value, p_value = pearsonr(filtered_df[selected_noise], filtered_df[dice_col])
       # r_squared = r_value ** 2
        
        # Add scatter trace for the model
        fig.add_trace(go.Scatter(
            x=filtered_df[selected_noise],
            y=filtered_df[dice_col],
            mode='markers',
            marker=dict(color=colors[model_name], size=8, opacity=0.7),
            name=model_name,
            customdata=filtered_df[['ct_name', selected_noise, dice_col]],
            hovertemplate=(
                "<b>ct_name:</b> %{customdata[0]}<br>" +
                f"<b>{selected_noise}:</b>" "%{customdata[1]:.2f}<br>" +
                f"<b>{dice_col}:</b>" "%{customdata[2]:.2f}<br>" +
                "<extra></extra>"
            )
        ))
        
        # Fit OLS linear regression
        X = filtered_df[selected_noise]
        Y = filtered_df[dice_col]
        X = sm.add_constant(X)  # Adds a constant term for the intercept
        model = sm.OLS(Y, X).fit()
        
        # Define x range for trendline
        x_min = filtered_df[selected_noise].min()
        x_max = filtered_df[selected_noise].max()
        x_trend = np.linspace(x_min, x_max, 100)
        X_trend = sm.add_constant(x_trend)
        y_trend = model.predict(X_trend)
        
        # Add trendline trace for the model
        fig.add_trace(go.Scatter(
            x=x_trend,
            y=y_trend,
            mode='lines',
            line=dict(color=colors[model_name], dash='dash'),
            name=f"{model_name} Trendline",
            showlegend=False
        ))
        
        # Create annotation text for the model
        annotation_text = (
            f"<b>{model_name}</b><br>"
            f"Pearson r = {r_value:.2f}<br>"
            f"p-value = {p_value:.2e}<br>"
         #   f"R² = {r_squared:.2f}"
        )
        
        # Position annotations based on model index
        if idx == 1:
            # First model annotation (top-left corner)
            annotations.append(dict(
                xref="paper", yref="paper",
                x=0.55, y=0.15,
                text=annotation_text,
                showarrow=False,
                font=dict(size=12, color=colors[model_name]),
                bgcolor="rgba(255,255,255,0.5)"
            ))
        elif idx == 2:
            # Second model annotation (top-right corner)
            annotations.append(dict(
                xref="paper", yref="paper",
                x=0.95, y=0.15,
                text=annotation_text,
                showarrow=False,
                font=dict(size=12, color=colors[model_name]),
                bgcolor="rgba(255,255,255,0.5)",
                xanchor='right'
            ))
    
    # Update layout for clarity and square shape
    fig.update_layout(
        title=f"Dice Scores vs {selected_noise} for AI Segmentation Models (Central 90% Noise)",
        xaxis_title= 'noise (HU)', #selected_noise,
        yaxis_title="Mean Dice Score",
        hovermode='closest',
        width=800,      # Adjust width as needed
        height=800,     # Adjust height as needed to make the plot square
        annotations=annotations
    )
    
    # Create output filename
    output_file = os.path.join(output_dir, f"Dice_vs_{selected_noise}_TwoModels_90percent.html")
    
    # Save the figure as an HTML file
    fig.write_html(output_file)
    print(f"Saved plot as {output_file}")
    
    # Display the figure with a specific renderer if needed
    fig.show(renderer="iframe")  # Use 'notebook', 'iframe', etc., based on your environment

# ==========================
# Execute the Plotting
# ==========================
plot_two_models_noise_vs_dice(df, selected_noise, models, colors, output_dir)
```
